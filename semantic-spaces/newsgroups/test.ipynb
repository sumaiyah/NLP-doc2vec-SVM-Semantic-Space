```python
import import_ipynb

import build_model, data_preprocessor, data_vis
from build_model import Model
```

## Get data

```python
raw_data, labels = data_preprocessor.get_dataset(n_samples_per_cat=500)
labels
```

```python
clean_parameters = dict(
    stem=True,
    rem_punc=True,
    stop=True,
    lemmatize=False
)

clean_param_str = str(clean_parameters)
clean_data = data_preprocessor.pre_process(raw_data, clean_parameters)
```

## Build Model

```python
model = Model(documents=clean_data,
              labels=labels,
              clean_param_str=clean_param_str, 
              model_name = 'model_2.model')
```

```python
document_vectors = model.infer_all_vectors()
```

## Visualise model

```python
data_vis.draw_tsne(model_name=model.model_name, doc_vectors=document_vectors, labels=model.labels)
```
